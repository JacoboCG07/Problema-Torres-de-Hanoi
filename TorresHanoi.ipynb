Funciones:

```python
import copy;

#Funcion que comprueba estados
def is_valid_state(s):
  assert type(s) == list

  def is_tower_valid(t): #Miramos si las torres son válidas
    for n in range(len(t)-1):
      if(t[n]>t[n+1]):
        return False
    return True

  for t in range(len(s)):#Miramos torre por torre si tienen los discos bien colocados
    if is_tower_valid(s[t]):
      pass
    else:
      return False
  return True

def states(s):
  for torre_quita in range(len(s)): #Bucle que recorre todas las torres
      for torre_añade in range(len(s)): #Bucle para que mire las otras dos torres
        ns=copy.deepcopy(s) #nuevo estado
        if torre_añade != torre_quita  and ns[torre_quita]: #Para que no la coloque en la misma torre y mira si la torre de donde va a quitar algo está vacía
          ns[torre_añade].insert(0,ns[torre_quita][0]) #Le añado a la ns[torre_añade] (Que será una torre) el primer elemento (disco) que hay en ns[torre_quita] (Que será otra torre)
          ns[torre_quita].pop(0) #Borro el primer elemento de ns[torre_quita](que será una torre) porque ya lo añadí a otra
          yield ns

```

Celda de prueba de funciones

```python
#Celda de prueba de las funciones de arriba
estado = [[], [], [], [5, 4, 3, 2, 1], []]

#Ejemplo de prueba
print("Primer estado", estado)
print("Estados que crea")
for s in states(estado):
  print(s)

print(is_valid_state(estado))
```

Generamos el estado inicial y el final

```python
estado_inicial = []
estado_final   = []

#Generamos las torres
torres = 3
c=1
while c<=torres:
  estado_inicial.append([])
  estado_final.append([])
  c+=1

#Generamos los discos
discos = 5
c=1
while c<=discos:
  estado_inicial[0].append(c)
  estado_final[len(estado_final)-1].append(c)
  c+=1
```

Definimos el algoritmo y mostramos el camino hasta la solución

```python
#Funcion recursividad

def busquedaProfundidad(actual, cerrados=[]):
    if actual==estado_final:
        return [actual]
    else:
        for ns in states(actual):
            if is_valid_state(ns) and ns not in cerrados:
                #print("estado valido: ",ns)
                cerrados.append(ns)
                r=busquedaProfundidad(ns,cerrados) #r = a la funcion busqueda en profundidad que devuelve una lista por eso añade
                if r:
                    r.insert(0,actual)
                    return r

        return []

print("Estado inicial",estado_inicial," y final ",estado_final)
solucion = busquedaProfundidad(estado_inicial)

print("Recorrido")
for s in solucion:
    print(s)
```
